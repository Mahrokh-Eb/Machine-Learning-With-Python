## Using Sklearn for transformation

#### Loading iris dataset

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
x, y = iris.data, iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
```

#### Mean and Standard Deviation of iris dataset

```python
print('mean : %s' %x_train.mean(axis=0))
print("standard deviation : %s" %x_train.std(axis=0) )
```

#### Standardization
#### 1- import

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

#### 2- fit ( )

```python
scaler.fit(x_train)  # getting Mu and Sigma
```

#### 3- transform ( )

```python
x_train_scaled = scaler.transform(x_train)
print("mean: %s"%x_train_scaled.mean(axis=0))
print("standard deviation: ", x_train_scaled.std(axis=0))
```

```python
x_test_scaled = scaler.transform(x_test)
print("mean: %s"%x_test_scaled.mean(axis=0))
print("standard deviation: ", x_test_scaled.std(axis=0))
```
